```python
#Generating a Pie chart with the distribution of CoVID-19 case types.
# Compute the total figures for each category of case (Deaths, Recovered, Active)
case_rate = df[['Deaths', 'Recovered', 'Active']].sum()
# Generate a pie chart to visualize the distribution of COVID-19 case types
plt.figure(figsize=(8, 8))  
# Set the figure size
plt.pie(case_rate, labels=case_rate.index,autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})  
# Set the chart title
plt.title('Distribution of COVID-19 Case Types')  
# Ensure the pie chart is a circle
plt.axis('equal')  
# Visualize the chart
plt.show()  
```

```python
# Produce a scatter plot to visualize the relationship between new cases and new deaths in COVID-19 cases.
# Constructing the figure size to 10x6 inches
plt.figure(figsize=(10, 6)) 

# Displaying a scatter plot to mapping 'New cases' to the x-axis and 'New deaths' to the y-axis
plt.scatter(df['New cases'], df['New deaths'], color='blue', alpha=0.6, edgecolor='black')  
# Adding a title to the plot with a specified font size
plt.title('New Cases vs New Deaths ', fontsize=16)

# Adding titles and font sizes to the x-axis and y-axis labels
plt.xlabel('New Cases', fontsize=13)
plt.ylabel('New Deaths', fontsize=13)

# Adding a grid to the plot for better readability
plt.grid(True)

# Adjust the layout to make sure all elements fit within the figure
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Choose essential columns for analysis in the COVID-19 dataset
data_attributes= ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 
                'New deaths','New recovered']
Relevant_data = df[data_attributes]

# Compute the correlation matrix
correlation_matrix = Relevant_data.corr()

# Compute the correlation matrix
correlation_matrix = Relevant_data.corr()

# Generate the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='cividis', fmt=".2f", 
            square=True, linewidths=0.1, linecolor='black')
plt.title('Correlation Heatmap of COVID-19 Data', fontsize=16)
plt.tight_layout()

# Display the heatmap
plt.show()
```
